```python
# Import dependencies:
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import statistics
import numpy as np
import pandas as pd
```

```python
# Load in csv
pyber = pd.read_csv("/Users/diegoperezo97/Documents/Tecnológico de Monterrey – Data Analytics Boot Camp/pyber-analysis/Module 5/Pyber Analysis/Resources/PyBer_ride_data.csv")
pyber.head()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber))
tick_locations = [value for value in x_axis]

# Plot the data.
pyber.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber["Month"])
plt.show()
```

```python
pyber.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber.plot(x="Month", y="Avg. Fare ($USD)", kind="bar")
plt.show()
```

```python
# Get the standard deviation of the values in the y-axis:
stdev = statistics.stdev(pyber["Avg. Fare ($USD)"])
stdev 
```

```python
# Create a vertical bar chart:
plt.bar(pyber["Month"], pyber["Avg. Fare ($USD)"], color="xkcd:sky blue", label="Avg. Fare ($ USD)", yerr=stdev, capsize=3)

# Create labels for the x and y axes:
plt.xlabel("Month")
plt.ylabel("Fare [$ USD]")

# Create a title:
plt.title("PyBer Fare by Month")

# Add the legend:
plt.legend()
```
